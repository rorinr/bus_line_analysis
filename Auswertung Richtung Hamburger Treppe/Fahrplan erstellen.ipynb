```python
import pandas as pd
pd.set_option('display.max_columns', 100)
pd.set_option('display.max_rows', 2500)
```

```python
Connection = pd.DataFrame()
df = pd.DataFrame()
Haltestellen = ["Hansastrasse", "Stralsunderstrasse", "Greifswalderstrasse", "Rostockerstrasse", "Bremerstrasse", "BethesdaKrankenhaus", "Friedhofskirche", "Marienstrasse", "Hoechsten", "Karlsplatz", "Hauptbahnhof", "Haubahn", "Wolkenburg", "Bendahlerstrasse", "Hesselnberg", "Christbusch", "FriedhofUnterbarmen", "Meckelstrasse", "ImSpringen", "Zeughausstrasse", "BarmenBahnhof", "AlterMarkt", "Landwehrplatz", "Leimbach", "Sedanstrasse"]
Haltestellen = list(Haltestelle for Haltestelle in reversed(Haltestellen))
names = ["Arrivetime", "Line", "Delay", "Goal", "Location", "Date", "Weekday", "Savetime"]
```

```python
#Liste mit allen Daten erstellen und als Spalte in Df
AllDates = set()
for Haltestelle in Haltestellen:  
    data = pd.read_csv("../Data und Mining/Data" + Haltestelle + ".csv", names = names)
    data.drop_duplicates(subset = ["Arrivetime", "Line", "Goal", "Location", "Date", "Weekday"], keep="last", inplace=True)
    data = data[(data.Line == 628)]
    data = data[data.Goal != "Wuppertal Sedanstraße"]
    data = data[data.Goal != "Wuppertal SedanstraÃŸe"]
    data = data[data.Goal != "Wuppertal Sedanstraße "]
    data = data[data.Goal != "Wuppertal Hbf"]
    data = data[data.Goal != "Wuppertal Hbf "]
    data = data[(data.Weekday != 5)]
    data = data[(data.Weekday != 6)]
    for date in data.Date.unique().tolist():
        if (len(data.loc[data.Date == date].Arrivetime) > 40):
            if (str(date) not in AllDates):
                AllDates.add(date)
AllDates = list(AllDates)
Connection["Date"] = AllDates * 41 #Anzahl der Fahrten an einem Tag
```

```python
#Fahrplan erstellen, zunächst ohne Verspötungen
for Haltestelle in Haltestellen:  
    data = pd.read_csv("../Data und Mining/Data" + Haltestelle + ".csv", names = names)
    data.drop_duplicates(subset = ["Arrivetime", "Line", "Goal", "Location", "Date", "Weekday"], keep="last", inplace=True)
    data = data[(data.Line == 628)]
    data = data[data.Goal != "Wuppertal Sedanstraße"]
    data = data[data.Goal != "Wuppertal Sedanstraße "]
    data = data[data.Goal != "Wuppertal SedanstraÃŸe"]
    data = data[data.Goal != "Wuppertal SedanstraÃŸe "]
    data = data[data.Goal != "Wuppertal Hbf"]
    data = data[data.Goal != "Wuppertal Hbf "]
    data = data[(data.Weekday != 5)]
    data = data[(data.Weekday != 6)]
    data.drop(["Weekday", "Goal", "Weekday", "Savetime", "Line", "Location"], axis = 1, inplace = True)
    
    #Um nur Wochentage zu bekommen, werden nur Tage mit mehr als 38 Fahrten ausgewertet
    actuall_times = []
    for date in data.Date.unique().tolist():
        if (len(data.loc[data.Date == date].Arrivetime) > 40):
            for i in data.loc[data.Date == date].Arrivetime:
                if i not in actuall_times:
                    actuall_times.append(i)
                    
    #In manchen Haltstellen befinden sich fehlerhafte Abfahrdaten, diese gilt es zu löschen
                    
    if (Haltestelle == "Sedanstrasse"):
        actuall_times = list(set(actuall_times).difference(set(['04:37', '05:07', '05:18'])))
        
    elif (Haltestelle == "Leimbach"):
        actuall_times = list(set(actuall_times).difference(set(['04:42', '05:12', '06:06', '19:06'])))
        
    elif (Haltestelle == "Landwehrplatz"):
        actuall_times = list(set(actuall_times).difference(set(['04:45', '05:15'])))
        
    elif (Haltestelle == "AlterMarkt"):
        actuall_times = list(set(actuall_times).difference(set(['04:48', '05:18'])))
        
    elif (Haltestelle == "BarmenBahnhof"):
        actuall_times = list(set(actuall_times).difference(set(['04:50', '05:20'])))

    elif (Haltestelle == "Zeughausstrasse"):
        actuall_times = list(set(actuall_times).difference(set(['04:52', '05:22'])))
        
    elif (Haltestelle == "ImSpringen"):
        actuall_times = list(set(actuall_times).difference(set(['04:53', '05:23'])))
    
    elif (Haltestelle == "Meckelstrasse"):
        actuall_times = list(set(actuall_times).difference(set(['04:54', '05:24'])))
        
    elif (Haltestelle == "Hesselnberg"):
        actuall_times = list(set(actuall_times).difference(set(['07:42', '08:42', '09:02', '13:02', '13:42', '16:22', '16:42', '19:02', '04:59', '05:29'])))

    elif (Haltestelle == "FriedhofUnterbarmen"):
        actuall_times = list(set(actuall_times).difference(set(['04:57', '05:27'])))
        
    elif (Haltestelle == "Christbusch"):
        actuall_times = list(set(actuall_times).difference(set(['04:58', '05:28'])))
        
    elif (Haltestelle == "Bendahlerstrasse"):
        actuall_times = list(set(actuall_times).difference(set(['05:00', '05:30'])))
        
    elif (Haltestelle == "Wolkenburg"):
        actuall_times = list(set(actuall_times).difference(set(['05:01', '05:31'])))
        
    elif (Haltestelle == "Haubahn"):
        actuall_times = list(set(actuall_times).difference(set(['05:03', '05:33'])))
        
    elif (Haltestelle == "Hauptbahnhof"):
        actuall_times = list(set(actuall_times).difference(set(['05:07', '05:37', '05:57', '06:17'])))
        
    elif (Haltestelle == "Karlsplatz"):
        actuall_times = list(set(actuall_times).difference(set(['05:12', '05:42', '06:02', '06:23'])))
    
    elif (Haltestelle == "Hoechsten"):
        actuall_times = list(set(actuall_times).difference(set(['05:13', '05:43', '06:03', '06:24'])))
        
    elif (Haltestelle == "Marienstrasse"):
        actuall_times = list(set(actuall_times).difference(set(['05:14', '05:44', '06:04', '06:25'])))
        
    elif (Haltestelle == "Friedhofskirche"):
        actuall_times = list(set(actuall_times).difference(set(['05:15', '05:45', '06:05', '06:26'])))
        
    elif (Haltestelle == "BethesdaKrankenhaus"):
        actuall_times = list(set(actuall_times).difference(set(['05:19', '05:49', '06:09', '06:30', '13:29', '15:49'])))

    elif (Haltestelle == "Bremerstrasse"):
        actuall_times = list(set(actuall_times).difference(set(['05:20', '05:50', '06:10', '06:31'])))
        
    elif (Haltestelle == "Rostockerstrasse"):
        actuall_times = list(set(actuall_times).difference(set(['05:21', '05:51', '06:11', '06:32'])))
        
    elif (Haltestelle == "Greifswalderstrasse"):
        actuall_times = list(set(actuall_times).difference(set(['05:22', '05:52', '06:12', '06:33'])))
        
    elif (Haltestelle == "Stralsunderstrasse"):
        actuall_times = list(set(actuall_times).difference(set(['05:23', '05:53', '06:13', '06:34'])))
        
    elif (Haltestelle == "Hansastrasse"):
        actuall_times = list(set(actuall_times).difference(set(['05:24', '05:54', '06:14', '06:35'])))
        
    actuall_times.sort()
    print((Haltestelle))
    Connection["Time" + Haltestelle] = actuall_times * len(AllDates)
    data["Time" + Haltestelle] = data.Arrivetime
    Connection = pd.merge(Connection, data, on = ["Date", "Time" + Haltestelle], how = "left")
    Connection.drop(["Arrivetime"], axis = 1, inplace = True)
    Connection = Connection.rename(columns = {"Delay": "Delay" + Haltestelle})
```

```python
Connection["Month"] = Connection["Date"].str[3:5]
Connection["Day"] = Connection["Date"].str[0:2]
Connection = Connection.sort_values(["Month", "Day", "TimeSedanstrasse"])
Connection.drop(["Month", "Day"], axis = 1, inplace = True)
Connection
```
